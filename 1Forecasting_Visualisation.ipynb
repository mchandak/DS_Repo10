#### Line plot

```python
# create a line plot
import pandas as pd
from matplotlib import pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
series = pd.read_csv('https://raw.githubusercontent.com/mchandak/DS_Repo10/main/daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True,squeeze=True)
series.head()
```

#### Histogram and Density Plots

```python
# series = read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
series.hist()
plt.show()
```

```python
# create a density plot
#series = read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
series.plot(kind='kde')
plt.show()
```

#### Box and Whisker Plots by Interval

```python
# create a boxplot of yearly data
from pandas import read_csv
from pandas import DataFrame
from pandas import Grouper
from matplotlib import pyplot
#series = read_csv('https://raw.githubusercontent.com/mchandak/DS_Repo10/main/daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True,squeeze=True)
groups = series.groupby(Grouper(freq='A'))     # freq='A' for grouping year wise
years = DataFrame()
for name, group in groups:
    years[name.year] = group.values
years.boxplot()
pyplot.show()
```

#### Lag plot

```python
# create a scatter plot
from pandas import read_csv
from pandas.plotting import lag_plot
# series = read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
lag_plot(series)
pyplot.show()
```

```python
# create an autocorrelation plot

from pandas import read_csv
pyplot.figure(figsize = (40,10))
from statsmodels.graphics.tsaplots import plot_acf
#series = read_csv('https://raw.githubusercontent.com/slmsshk/DataSet/main/daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
plot_acf(series,lags=90)
plt.show()
```
