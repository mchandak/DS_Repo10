#### IsolationForest

```python
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mchandak/DS_Repo10/main/Data/iris2.csv")
df
```

```python
data_encoded1=pd.get_dummies(df)
data_encoded1
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded1)
```

```python
# predictions
y_pred_outliers = clf.predict(data_encoded1)
```

```python
#-1 for outliers and 1 for inliers.
y_pred_outliers
```

```python
## Let us add a new data point which is outlier
data_encoded1.loc[150]=[20,40,30,50,1,0,0]
```

```python
data_encoded1
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded1)
# predictions
y_pred_outliers = clf.predict(data_encoded1)
y_pred_outliers
```

```python
data_encoded1['scores']=clf.decision_function(data_encoded1)
data_encoded1
```

```python
data_encoded1['anomaly']=clf.predict(data_encoded1.iloc[:,0:7])
```

```python
data_encoded1
```

```python
#Print the outlier data points
data_encoded1[data_encoded1['anomaly']==-1]
```
