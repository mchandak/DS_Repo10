```python
# from pandas import read_csv
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.manifold import TSNE 
from bioinfokit.visuz import cluster
```

```python
# load data
dataframe = pd.read_csv("https://raw.githubusercontent.com/mchandak/DS_Repo10/main/Data/TSNE_data.csv")
```

```python
dataframe.head()
```

```python
# Split-out validation dataset
array = dataframe.values
# separate array into input and output components
X = array[:,1:]
Y = array[:,0]
```

```python
dataframe
```

```python
data_tsne = TSNE(n_components=2,random_state=10).fit_transform(X)
# cluster.tsneplot(score=data_tsne)
```

```python
# get a list of categories
color_class = dataframe['diagnosis'].to_numpy()
cluster.tsneplot(score=data_tsne, colorlist=color_class, legendpos='upper right')
# plt.show()
#Plot will be stored in the default directory
```
